# Naive Bayes 

<b> Bag of words </b>

- count the word frequency of a file that has a bunch of text, represent them as a set and have a count next to them 
- position doesn't matter 

"happy" : 5 
"anyone" : 2
"who" : 2
"is" : 2

<b>BOW Features</b> 
- for each pair of words, we estimate the probability of word given a class 



- count the number of times the word wi occurs in document cj then divide it by the sum of the counts that are associated in the count cj

# Example 




1. Total words V = [Chinese, Beijing, Shanghai, Macao, Tokyo, Japan] || |V| = 6
2. prior probability 
    - P(c) = 3/4
    - P(j) = 1/4 
3. Estimate the conditional probability 

<b>Class C MLE</b>
- P(chinese|C) = 5/8
- P(beijing|c) = 1/8
- p(shanghai|c) = 1/8
- p(Macao|c) = 1/8
- P(tokyo|c) = 0/8
- p(japan|c) = 0/8
 
<b>Class C MLE & Add One smoothing </b>



|v| = 6

- p(chinese|c) = 6/14
- P(beijing|c) = 2/8
- p(shanghai|c) = 2/14
- p(Macao|c) = 2/14
- P(tokyo|c) = 1/14
- p(japan|c) = 1/14

Class J MLE
- P(tokyo|j) = 1/3
- p(japan|j) = 1/3
- p(chinese|j) = 1/3
- P(beijing|j) =  0/3
- p(shanghai|j) =  0/3
- p(Macao|j) =  0/3

<b>Class J MLE & Add One smoothing </b>



|v| = 6

- p(chinese|j) = 2/9
- P(beijing|j) = 0/9
- p(shanghai|j) = 0/9
- p(Macao|j) = 0/9
- P(tokyo|j) = 2/9
- p(japan|j) = 2/9

<b> Conditional Probabilities of Class C </b>


// order doesn't matter, but which words occur in the document and how many times have I seen it? 

- P(chinese|c) = (5+1) / (8+6) = 6/14 = 3/7
- P(tokyo|c) = (0+1)/(8+6)
- p(japan|c) = (0+1)/(8+6)
- P(beijing|c) =  (1+1)/(8+6)
- p(shanghai|c) =  (1+1)/(8+6)
- p(Macao|c) = (1+1)/(8+6)


* Denominator 8 represents the number of tokens in class C 
* Denominator 6 represents the number of types |v| in the vocabulary   
* numerator 5 the number of occurences of that type (chinese) in that class 
* numerator 1 is for add one smoothing
