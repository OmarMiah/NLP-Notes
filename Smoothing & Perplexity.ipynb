# Evaluating Language Models

- Perplexity: How well our Language model fits our test data 

- We get test data
- n sentences -> the probability of each sentence is p(si) 


underflow - when you multiply the probabilty of each and get a number super small.

To fix this problem you have to take the log of each probability and sum them up.

<b> Perplexity </b>

preplexity = 2^-l

l = average log probabablity word by word 
m = number of tokens in the test data 

So taking our test data:

log(p(sentence_i)) 


<b> log(p(unigram_estimate*products)) </b>

<b> average log probability </b> 

    log(p(sentence_i) = p(w_1) X p(w2) x p(w3) x ... xp(wn))
    = log(p(w_1)) + log(p(w2)) + log(p(w3)) + ... + log(p(wn))/ n Tokens
    
Then we average it over the total number of tokens 

The close the probability to 0 the greater the negative number will be. We have a negative in perplexity, we will get negative numbers. If the probabilty values are negative numbers, the low probability will have very large negative numbers. 

<b> High perplexity corresponds to a bad language model. </b>

# Perplexity and N-Gram order

- Bigram - better fit as they condition on the prior word 
- Trigram -> fits the best since there is a large amount of training data 

# Sparse Data Problems


<b> Maximum likelihood</b>

    - p(wi) = c(wi)/n 
    
- in large corpus a lof of n-grams won't occur and will have a probability of 0 under the MLE estimate 

bigrams and trigrams are more likely to get a probability of zero 

<b> Solution to sparse data (MLE POOR ESTIMATES) </b>

- smoothing<br>

    <b> Add one smoothing (UNIGRAMS)</b>
        - we under count the number of our n-grams .
        - Maximum likelihood
        - p(wi) = c(wi)/n 
        - adjusted count
            - c*(wi) = c(wi) + 1 // adding one to each count
        - then our addone smoothing probab will be
            - p*(wi) = (c(wi)+1)/(N+V)
            - we add v because is the size of the vocabulary, now we have to add one to training data 
            
    <b> Add one smoothing (BIGRAMS)</b>
        - we under count the number of our n-grams .
        - Maximum likelihood
        - p(wi|wi-1) = c(wi-1)/c(wi-1) 
        - adjusted count
            - c*(wi|wi-1) = c(wi|wi-1) + 1 // adding one to each count
        - then our addone smoothing probab will be
            - p*(wi|wi-1) = (c(wi-1, wi)+1)/ c(wi-1)+(V)
            - we add v because is the size of the vocabulary, now we have to add one to training data 

# Add-One Smoothing Example

<b> vocab </b> 

- would = v = {i, like, to, here, the, would}

<b> Bigrams </b> 

<b> MLE ESTIMATES </b><br>
&emsp;&emsp; would i 0     &emsp;        p(i|would) = 0/8 <br>
&emsp;&emsp; would like 5  &emsp;     p(like|would) = 5/8 <br>
&emsp;&emsp; would to 2    &emsp;       p(to|would) = 2/8 <br> 
&emsp;&emsp; would here 1  &emsp;     p(here|would) = 1/8 <br>
&emsp;&emsp; would the 0   &emsp;      p(the|would) = 0/8 <br>
&emsp;&emsp; would would 0 &emsp;    p(would|would) = 0/8<br>

- <b> note</b>: We got 8 here because we took the some of the estimates and used it in the denominator 

<b> Add-One Smoothing </b>
&emsp;&emsp; would i 1      &emsp;  p(i|would) = 1/(8+6) <br>
&emsp;&emsp; would like 6   &emsp;  p(like|would) = 6/(8+6) <br>
&emsp;&emsp; would to 3     &emsp;  p(to|would) = 3/(8+6) <br>
&emsp;&emsp; would here 2   &emsp;  p(here|would) = 2/(8+6) <br>
&emsp;&emsp; would the 1    &emsp;  p(the|would) = 1/(8+6) <br>
&emsp;&emsp; would would 1  &emsp;  p(would|would) = 1/(8+6)<br>

- <b> note</b>: We got 14 here because we took the some of the estimates and used it in the denominator 


Add-One smoothing allows us to take away some probability mass from observed events and re-assign it to "unseen" events
